## 2. Conversiones y Mapping

Importación de las librerias

```python
import pandas as pd
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
import lightgbm as lgb
from sklearn.metrics import accuracy_score

```

Conversion en Columnas Si y No en Binarios

```python
z_test.F_TIENEINTERNET.value_counts()
```

```python
# Conversion en la columna F_TIENEINTERNET
map_binario = {'Si': 1, 'No': 0}
z_train['F_TIENEINTERNET'] = z_train['F_TIENEINTERNET'].map(map_binario)
z_test['F_TIENEINTERNET'] = z_test['F_TIENEINTERNET'].map(map_binario)
```

```python
z_test.F_TIENEINTERNET.value_counts()
```

```python
z_test.E_PAGOMATRICULAPROPIO.value_counts()
```

```python
# Conversion en la columna E_PAGOMATRICULAPROPIO
map_binario = {'Si': 1, 'No': 0}
z_train['E_PAGOMATRICULAPROPIO'] = z_train['E_PAGOMATRICULAPROPIO'].map(map_binario)
z_test['E_PAGOMATRICULAPROPIO'] = z_test['E_PAGOMATRICULAPROPIO'].map(map_binario)
```

Conversión de valores categoricos a numéricos

```python
z_train.E_VALORMATRICULAUNIVERSIDAD.value_counts()
```

```python
# Conversion en columna VALORMATRICULAUNIVERSIDAD (promedio)
map_promedio = {
    "No pagó matrícula": 0.0, "Entre 5.5 millones y menos de 7 millones": 6.25, "Más de 7 millones": 7.5, "Entre 4 millones y menos de 5.5 millones": 4.75, "Entre 500 mil y menos de 1 millón": 0.75,
    "Menos de 500 mil": 0.25, "Entre 2.5 millones y menos de 4 millones": 3.25, "Entre 1 millón y menos de 2.5 millones": 1.75

}
z_train['E_VALORMATRICULAUNIVERSIDAD'] = z_train['E_VALORMATRICULAUNIVERSIDAD'].map(map_promedio)
z_test['E_VALORMATRICULAUNIVERSIDAD'] = z_test['E_VALORMATRICULAUNIVERSIDAD'].map(map_promedio)

```

```python
z_test.E_VALORMATRICULAUNIVERSIDAD.value_counts()
```

```python
z_test.F_ESTRATOVIVIENDA.value_counts()
```

```python
# Conversion en la columna F_ESTRATOVIVIENDA
map_estrato = {
   "Estrato 1": 1, "Estrato 2": 2, "Estrato 3": 3, "Estrato 4": 4, "Estrato 5": 5, "Estrato 6": 6, "Sin Estrato": 0
}
z_test['F_ESTRATOVIVIENDA'] = z_test['F_ESTRATOVIVIENDA'].map(map_estrato)
z_train['F_ESTRATOVIVIENDA'] = z_train['F_ESTRATOVIVIENDA'].map(map_estrato)
```

```python
z_train.RENDIMIENTO_GLOBAL.value_counts()
```

```python
# Conversion columna RENDIMIENTO_GLOBAL
map_rendimiento = {
    'bajo': 0, 'medio-bajo': 1, 'medio-alto': 2, 'alto': 3
}
z_train['RENDIMIENTO_GLOBAL'] = z_train['RENDIMIENTO_GLOBAL'].map(map_rendimiento)
```

```python
z_train.RENDIMIENTO_GLOBAL.value_counts()
```

## 3. Separación y variable objetivo

```python
# Separacion de caracteristicas (X)
X_train = z_train.drop(columns=['ID', 'RENDIMIENTO_GLOBAL'])

# Variable Objetivo (y)
y_train = z_train['RENDIMIENTO_GLOBAL']
X_test = z_test.drop(columns=['ID'])
```

Preprocesamiento Pipeline para datos númericos y categoricos

```python
preprocessor = ColumnTransformer(
    transformers=[
        # Columnas numéricas y StandardScaler para la normalización de datos númericos
        ('num', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', StandardScaler())
        ]), X_train.select_dtypes(include=['float64','int64']).columns),
        # Columnas Categoricas y OneHotEncoder
        ('cats', OneHotEncoder(handle_unknown='ignore'),
         X_train.select_dtypes(include=['object']).columns)
    ]
)
```

## 4. Modelo LightGBM

Creación de Pipeline

```python
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', lgb.LGBMClassifier(
        objective='multiclass', n_estimators=1000, learning_rate=0.02, num_leaves=64, min_child_samples=50,
        feature_fraction=0.8, bagging_fraction=0.8, bagging_freq=5, random_state=42, n_jobs=-1
      )
    )
    ]
)
```

Entrenamiento de Modelo

```python
# Validación para la evaluación del modelo especialmente rendimiento
cv = StratifiedKFold(n_splits=4, shuffle=True, random_state=42)
cv_scores = cross_val_score(pipeline, X_train, y_train, cv=cv, scoring='accuracy')

print("LGBM:", cv_scores.mean(), "+/-", cv_scores.std())
```

```python
# Entrenamos el modelo
pipeline.fit(X_train, y_train)
```

## Predicciones y Creación de Archivo

Prediccion en el conjunto de pruebas

```python
# Prediccion
predictions = pipeline.predict(X_test)
```

```python
predictions
```

```python
# Conversion de las predicciones
map_rendimiento2 = {
    0: 'bajo', 1: 'medio-bajo', 2: 'medio-alto', 3: 'alto'
 }
predictions_labels = [map_rendimiento2[pred] for pred in predictions]
```

Archivo de Presentación

```python
# Creación del archivo de envio con las predicciones
submission_df = pd.DataFrame({
    "ID": z_test['ID'],
    "RENDIMIENTO_GLOBAL": predictions_labels
})

submission_df.to_csv('submission.csv', index=False)

```
